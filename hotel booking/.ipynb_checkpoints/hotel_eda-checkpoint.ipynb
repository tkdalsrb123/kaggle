## 데이터 설명
https://www.kaggle.com/datasets/jessemostipak/hotel-booking-demand?datasetId=511638&sortBy=voteCount

- is_canceled: 예약 취소(1), 아닌지(0)    
- lead_time: 예약이 입력된 날짜와 도착 날짜 사이에 경과된 일수   
- stays_in_weekend_nights: 손님이 호텔에 숙박했거나 숙박을 예약한 주말 밤(토요일 또는 일요일)의 수   
- stays_in_week_nights: 손님이 호텔에 숙박했거나 숙박을 예약한 주중 밤 수(월요일~금요일)   
- market_segment: 시장 세그먼트 지정입니다. 범주에서 "TA"라는 용어는 "여행사"를 의미하고 "TO"는 "투어 운영자"를 의미합니다    
- distribution_channel: 예약 유통 채널   
- is_repeated_guest: 예약 이름이 반복되는 게스트인지(1), 아닌지(0)   
- agent: 예약한 여행사의 id   



| 목표: hotel booking cancellation eda and prediction   
종속변수: is_canceled
